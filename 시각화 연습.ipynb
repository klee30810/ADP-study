## 수치형 데이터 시각화
- 일정한 범위 내에서 어떻게 분포되어있는지 확인 필요 -> 어떻게 변환 및 해석할 지 결정 가능

```python
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

titanic = sns.load_dataset('titanic')

mpl.rc('font', size=10)
plt.figure(figsize=(8,6))
grid = gridspec.GridSpec(2,1)
plt.subplots_adjust(wspace=0.45, hspace=0.3)

ax = plt.subplot(grid[0])
sns.histplot(data=titanic, x='age', hue='alive', multiple='stack', ax=ax);
ax = plt.subplot(grid[1])
sns.kdeplot(data=titanic, x='age', ax=ax);
sns.rugplot(data=titanic, x='age', ax=ax); # 특정 단일 피쳐의 분포 표시
```

## 범주형 데이터 시각화

```python
mpl.rc('font', size=8)
plt.figure(figsize=(8,6))
grid = gridspec.GridSpec(3,2)
plt.subplots_adjust(wspace=0.2, hspace=0.3)

ax = plt.subplot(grid[0])
sns.barplot(x='class', y='fare', data=titanic); # 범주형 데이터에 따른 수치형 데이터의 평균과 신뢰구간
ax = plt.subplot(grid[1])
sns.pointplot(x='class', y='fare', data=titanic); # 같은 내용이나 여러 그래프 복합 비교
ax = plt.subplot(grid[2])
sns.boxplot(x='class', y='age', data=titanic); # 구체적 요약 수치
ax = plt.subplot(grid[3])
sns.violinplot(x='class', y='age', hue='sex', data=titanic, split=True); # 박스플롯 + 커널밀도추정 함수 : 수치형 데이터의 전체적 분포 양상
ax = plt.subplot(grid[4])
sns.countplot(x='class', data=titanic); # 타겟 값 분포 확인
ax = plt.subplot(grid[5])
x = [10,60,30]; labels = ['a', 'b', 'c'];
plt.pie(x=x, labels=labels, autopct= '%.1f%%'); # 범주형 데이터 별 비율
```

## 데이터 관계 시각화
- 히트맵, 라인플롯, 산점도, 회귀선 산점도

```python
mpl.rc('font', size=8)
plt.figure(figsize=(8,6))
grid = gridspec.GridSpec(2,2)
plt.subplots_adjust(wspace=0.2, hspace=0.3)

ax = plt.subplot(grid[0])
flights = sns.load_dataset('flights')
flights_pivot = flights.pivot(index='month', columns='year', values='passengers') # 연도의 월별 데이터 수
sns.heatmap(data=flights_pivot);
ax = plt.subplot(grid[1])
sns.lineplot(x='year', y='passengers', data=flights); # 두 수치형 데이터 사이의 관계
ax = plt.subplot(grid[2])
sns.scatterplot(x='year', y='passengers', hue='month', data=flights); # 두 데이터의 관계
ax = plt.subplot(grid[3])
sns.regplot(x='year', y='passengers', ci=95, data=flights); # 상관관계 파악
```
