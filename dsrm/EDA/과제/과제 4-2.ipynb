```python
## 과제 4-2 


import pandas as pd 
from sklearn.model_selection import train_test_split
X = pd.read_csv('../../data/빅분기_X_train.csv',encoding='cp949')
y = pd.read_csv('../../data/빅분기_y_train.csv',encoding='cp949')
```

```python
X
```

```python
# 빅분기 X_train, 빅분기 y_train 데이터의 cust_id를 인덱스로 설정하세요 
X.set_index('cust_id', inplace=True)
y.set_index('cust_id', inplace=True)
```

```python
# X_train과 y_train데이터를 y값을 기준으로 층화추출하여 7:3으로 데이터를 나누세요 (X_train, X_val, y_train,y_val)
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.3, stratify =y, random_state=2022)
```

```python
y_val.value_counts()
```

```python
# X_train과 X_val의 수치형 변수들만 StandardScaler를 이용하여 표준화 시키세요. 
from sklearn.preprocessing import StandardScaler

cols = [['총구매액', '최대구매액', '환불금액', '내점일수','내점당구매건수', '주말방문비율', '구매주기']]
X_train_num = X_train[['총구매액', '최대구매액', '환불금액', '내점일수','내점당구매건수', '주말방문비율', '구매주기']]
X_val_num = X_val[['총구매액', '최대구매액', '환불금액', '내점일수','내점당구매건수', '주말방문비율', '구매주기']]
scaler = StandardScaler()

X_train_num = scaler.fit_transform(X_train_num)
X_val_num = scaler.fit_transform(X_val_num)

X_train_num = pd.DataFrame(X_train_num, columns = cols)
X_val_num = pd.DataFrame(X_val_num, columns = cols)

X_train_num
```
