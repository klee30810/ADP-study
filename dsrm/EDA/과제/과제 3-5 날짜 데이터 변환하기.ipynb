# 날짜 데이터 전처리 과제 
- 데이터 : "6.세종시_오피스텔(매매)_실거래가.csv"

```python
import pandas as pd 
df_sejong = pd.read_csv("../../data/6.세종시_오피스텔(매매)_실거래가.csv",encoding='utf8')
```

```python
df_sejong
```

```python
### 아래와 같이 계약년월을 datetime으로 변경하세요. 
df_sejong['계약년월'] = pd.to_datetime(df_sejong['계약년월'].astype(str),format="%Y%m") 
print(df_sejong)
print(df_sejong.info())
```

```python
## 계약년월을 아래와 같이 문자열로 다시 변경하세요. 
df_sejong['계약년월'] = df_sejong['계약년월'].dt.strftime("%Y_%m_%d")
df_sejong
```

```python
df_sejong.info()
```

```python
df_sejong['계약월'] = list(map(lambda x: x[1],df_sejong['계약년월'].str.split("_")))
df_sejong
```

```python
## 6월 데이터만 추출해보세요 
df_sejong.loc[df_sejong['계약월']=='06'].head(5)
```

```python
## 날짜순으로 데이터를 정렬해보세요 
df_sejong.sort_values(by='계약년월')
```
