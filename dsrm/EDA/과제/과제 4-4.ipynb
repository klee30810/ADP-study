```python
import pandas as pd
df = pd.read_csv('../../data/titanic.csv')
df
```

```python
## 1. 타이타닉 데이터의 결측치를 확인하세요 
df.info()
```

age : 177
embarked : 2
deck : 688
embark_town : 2

```python
## 2. age의 결측값은 중위 값으로 대체하세요. 
df['age'].fillna(df['age'].median(),inplace=True)
df
```

```python
df.info()
```

```python
df['embark_town'].mode()
```

```python
## 3. embark_town의 결측 값은 최빈값으로 대체하세요. 
df['embark_town'].fillna(df['embark_town'].mode(),inplace=True)
df
```

```python
df.info()
```

```python
## 4. embarked 결측치는 embark_town의 첫 글자로 대체하세요 
df['embarked'].fillna(df['embark_town'][0], inplace=True)
df
```

```python
df.info()
```

```python
## 5. df_t내에 범주형 변수(embark_town,class) 를 원 핫 인코딩하여 더미변수로 만드세요 
df = pd.get_dummies(data=df, columns=['embark_town', 'class'])
df
```
