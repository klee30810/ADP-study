```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

mpl.rc('font', family='AppleGothic')
```

# 2. 표와 그림을 통한 자료의 요약
## 범주형 자료의 요약

```python
# 3.1

data = {'단과대학' : ['공과대', '자연대', '경영대', '사법대', '법대', '사회대','가정대'], '학생수':[1300,300,500,300,200,400,170], \
    '학과수': [20,6,5,6,3,8,3]}
df = pd.DataFrame.from_dict(data=data, orient='columns')
df['상대비율'] = [x/np.sum(df['학생수']) for x in df['학생수']]

plt.pie(x=df['학생수'], labels=df['단과대학'], autopct= '%.1f%%'); # 범주형 데이터 별 비율

```

```python
sns.barplot(x='단과대학', y='상대비율', data=df);
```
