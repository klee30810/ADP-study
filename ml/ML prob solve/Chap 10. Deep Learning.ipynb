### 10.1.3 활성화함수
- 시그모이드 : S자 곡선 그리는 수학 함수
- ReLu : max(0,x)

### 10.1.4 경사하강법
- 손실함수 : 모델 예측값과 실젯값의 차이, RMSE & Cross Entropy
- 경사하강법 : 기울기 구함 -> 아래방향 일정거리 이동 -> 손실함수 최소 될 때까지(기울기 0) 반복 
    - 이동량 : learning rate $W=W-\eta\frac{\partial L}{\partial W} $
    
### 10.1.5 순전파와 역전파
- 순전파를 진행하여 입력값으로부터 출력값(타겟값)을 계산 -> 순전파로 구한 타깃 예측값과 실제 타깃값의 차이를 구함 -> 손실값을 줄이는 방향으로 가중치의 기울기 구함 -> 기울기를 바탕으로 가중치 갱신(출력-입력층 방향으로 차례로 가중치 갱신_ -> 갱신된 가중치 바탕으로 단계 반복

## 10.2 합성곱 신경망(CNN)

### 10.2.1 합성곱 계층
- 합성곱 : 2차원 데이터 일정 영역 내 값들을 하나의 값으로 압축하는 연상, 중요한 특징만 추출하는 효과
- 필터 : 입력 데이터에서 특정한 특성을 필터링하는 역할
- 특성맵 : 합성곱 연산의 결과

### 10.2.2 패딩과 스트라이드
- 패딩 : 입력 데이터 주변을 특정 값으로 채우는 것, 출력 특성맵과 입력 데이터 크기를 동일하게 유지하기 위함
- 스트라이드 : 합성곱 연산 시, 한번에 이동하는 간격
$$ N_{out} = [\frac{N_{in}+2P-K}{S}] + 1$$
- K : 필터크기, P: 패딩크기, S: 스트라이드 크기, []: 내림 기호

### 10.2.3 풀링
- 풀링 : 특성 맵 크기를 줄여 이미지의 요약정보 추출
    - 합성곱은 필터의 패턴과 유사한 영역추출, 풀링은 최댓값이나 평균을 가져와 요약정보를 구하므로 필터가 필요없음
    - 목적 : 특성 맵 크기를 줄여 연산속도 빠르게 & 이미지에서 물체의 위치가 바뀌어도 같은 물체로 인식하기 위함 (특정 영역의 대푯값을 사용하므로 위치가 약간 변해도 같은 물체로 판단, 위치불변성)
    - 최대 풀링 : 가장 뚜렷한 부분 추출
    - 평균 풀링 : 특정 영역의 평균적 특징 추출
    - 풀링과 스트라이드의 크기는 일반적으로 같은 값
$$ N_{out} = [\frac{N_{in}+2P-K}{S}] + 1 = N_{out} = [\frac{N_{in}+2*0-K}{S}] + 1 = \frac{N_{in}}{K} $$

### 10.2.4 전결합
- fully connected layer (dense layer) : 이전 계층의 모든 노드 각각이 다음 계층의 노드 전부와 연결된 결합
    - 평탄화 : 마지막 부분에서 구현하여, 특정한 레이블로 분류하는 역할

### 10.2.5 합성곱 신경망 전체 구조
- 합성곱 계층 -> 풀링 계층 -> 전결합 계층    

## 10.3 성능 향상을 위한 딥러닝 알고리즘

### 10.3.1 드롭아웃
- 과대적합을 방지하기 위하여, 신경망 훈련 과정에서 무작위로 일부 뉴런 제외
- 매 훈련 이터레이션마다 새롭게 적용 => 조합이 달라지므로 훈련 이터레이션마다 서로 다른 신경망 훈련 후 앙상블 학습과 유사
- 훈련 단계에서만 적용하고, 검증이나 예측 단계에서는 훈련된 뉴런을 모두 사용해야 하므로 적용하지 않음

### 10.3.2 배치 정규화
- 과대적합 방지와 훈련 속도 향상
- 내부 공변량 변화(Internal Covariate Shift) : 신경층마다 입력 데이터 분포가 다른 현상 => **계층 간 데이터 분포의 편차를 줄이는 작업 필요**
1) 정규화 : 미니배치를 평균이 0, 분산이 1이 되게 정규화
$$ \hat{x_i}=\frac{x_i-\mu_B}{\sqrt{\sigma_B^2+\epsilon}}$$
2) 스케일 조정 및 이동 $z_i=\gamma \hat{x_i} +\beta $
    - gamma : 스케일 파라미터 / beta : 이동 파라미터
    - 정규화 하면 대부분 0근처로 값이 모이므로 활성화 함수 지날 시 선형성에 빠질 수 있음. 
    
### 10.3.3 옵티마이저
- 신경망의 최적 가중치를 찾아주는 알고리즘
    - 모멘텀 : 이전 단계의 진행방향을 기억하여 일정 비율로 현 단계에 반영
    - Adagrad : 최적점에 도달할 수록 학습률을 낮추도록, adaptive learning rate, 개별 파라미터에 적응하며 학습률 갱신
    - RMSProp : 최근 기울기만 고려해 학습률 낮춰, 훈련을 오래 지속해도 학습률이 0에 수렴하지 않음
    - Adam : 모멘텀 + RMSProp
    
### 10.3.4 전이학습
- 한 영역에서 사전 훈련된 모델에 약간의 추가학습을 더해 유사한 다른 영역에서 활용
- 일부 계층만 훈련시켜도 유용할 수도 : 신경망의 각 계층이 고유한 역할을 담당할 수 있으므로 : 선, 패턴, 형상 추출은 대상마다 상관이 없는데 말과 낙타 데이터를 구분하기 위한 분류망만 추가 학습
